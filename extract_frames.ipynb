```python
import io
import os
import subprocess
import glob

def extract_frames(video, frame_rate):
    command = 'ffmpeg -i ./phone/{0}.avi -vf fps={1} ./Cabin_Frame/{2}/frame%d.jpg'.format(video, frame_rate, video)
    subprocess.call(command,shell=True)

video_dir = './phone'
video_list = glob.glob(os.path.join(video_dir,'Cabin*.avi'))
for video_path in video_list:
    video = os.path.split(video_path)[-1]
    video_name = os.path.splitext(video)[0]
    if not os.path.exists('./Cabin_Frame/{0}'.format(video_name)):
        os.mkdir('./Cabin_Frame/{0}'.format(video_name))
    extract_frames(video_name, 2)
```
